## Countries with highest population in year 2022

```python
top_pop2022=data.groupby('Country')["POP2022"].sum().reset_index().sort_values("POP2022",ascending=False)
```

```python
top_pop2022  # Countries with top  population 
```

## Continent with highest population in year 2022

```python
top_con2022=data.groupby('Continent')["POP2022"].sum().reset_index().sort_values("POP2022",ascending=False)
```

```python
top_con2022
```

##  Countries with highest population percentage worldwide in year 2022


```python
top_per_2022=data.groupby('Country')["World_Population_Percentage"].sum().reset_index().sort_values("World_Population_Percentage",ascending=False)
```

```python
top_per_2022
```

##  Continent with highest population percentage worldwide in year 2022


```python
top_perCoN_2022=data.groupby('Continent')["World_Population_Percentage"].sum().reset_index().sort_values("World_Population_Percentage",ascending=False)
```

```python
top_perCoN_2022
```

## Difference between 1970 and 2022 population continent wise

```python
data['diff_PoP'] = ((data.POP2022) - (data.POP1970))
```

```python
data['diff_PoP']
```

```python
Diff_Con_2022=data.groupby('Continent')["diff_PoP"].sum().reset_index().sort_values("diff_PoP",ascending=False)
```
