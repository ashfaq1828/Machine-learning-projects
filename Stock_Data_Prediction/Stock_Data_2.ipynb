## Save the Expanded button data

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import StaleElementReferenceException
from bs4 import BeautifulSoup
import pandas as pd

# Set the path to chromedriver.exe
chromedriver_path = "C:/Users/AshfaqAli/Documents/Azure Data engineering/Selenium/new/chromedriver-win64/chromedriver.exe"

# Set up the Chrome service
service = Service(chromedriver_path)

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(service=service)

# Open the URL in the browser
driver.get("https://www.screener.in/company/IRFC/")

# Find all elements with the class 'blue-icon'
elements = driver.find_elements(By.CLASS_NAME, 'blue-icon')

List_df_names = ['Peer Comparision','Quarterly Results  ', 'Profit & Loss', 'Compounded Sales Growth', 'Compounded Profit Growth',
                 'Stock Price CAGR', 'Return on Equity', 'Balance Sheet', 'Cash Flows', 'Ratios', 'Shareholding Pattern', 'duplicate_Shareholding Pattern']

# Click on the parent button of each element
for element in elements:
    try:
        WebDriverWait(driver, 50).until(EC.element_to_be_clickable((By.XPATH, "//button[contains(@onclick, 'showSchedule')]"))).click()
    except StaleElementReferenceException:
        continue

# Extract HTML content after clicking elements
html_code = driver.page_source

# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(html_code, 'html.parser')
tables = soup.find_all('table')
dataframes = []

for idx, table in enumerate(tables):
    df = pd.read_html(str(table))[0]
    dataframes.append(df)

    # Check if the index is within the range of List_df_names
    if idx < len(List_df_names):
        # Assign the name to the DataFrame
        df_name = List_df_names[idx]
        print(f"{df_name} DataFrame:\n{df}\n\n")
        
        # Save each DataFrame to a CSV file with the corresponding name
        df.to_csv(f"{df_name}.csv", index=False)

# Quit the driver after processing
# driver.quit()
```

## Read the CSV file

#### 1 Peer Comparision

```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\AshfaqAli\Documents\ML Learning\CSV_Files_Expanded\Peer Comparision.csv")

# Create an empty dictionary to store the formatted data
formatted_data = {}

# Iterate through the rows of the DataFrame
for index, row in df.iterrows():
    name = row['Name']
    formatted_data[f"{name}_CMP  Rs."] = row['CMP  Rs.']
    formatted_data[f"{name}_P/E"] = row['P/E']
    formatted_data[f"{name}_Mar Cap  Rs.Cr."] = row['Mar Cap  Rs.Cr.']
    formatted_data[f"{name}_Div Yld  %"] = row['Div Yld  %']
    formatted_data[f"{name}_NP Qtr  Rs.Cr."] = row['NP Qtr  Rs.Cr.']
    formatted_data[f"{name}_Qtr Profit Var  %"] = row['Qtr Profit Var  %']
    formatted_data[f"{name}_Sales Qtr  Rs.Cr."] = row['Sales Qtr  Rs.Cr.']
    formatted_data[f"{name}_Qtr Sales Var  %"] = row['Qtr Sales Var  %']
    formatted_data[f"{name}_ROCE  %"] = row['ROCE  %']


df = pd.DataFrame.from_dict(formatted_data, orient="index")
df = df.T

csv_file_path=r"C:\Users\AshfaqAli\Documents\ML Learning\Transpose Data\Peer Comparision\my_data_Peer_Comparision.csv"


df.to_csv(csv_file_path, index=False)



```
