## Profit & Loss

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.screener.in/company/IRFC/'  # Replace 'YOUR_URL_HERE' with the actual URL

# Fetch the webpage content
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the section with the specified class
    profit_loss_section = soup.find('section', {'id': 'profit-loss', 'class': 'card card-large'})

    # Find the table within the section
    table = profit_loss_section.find('table', {'class': 'data-table responsive-text-nowrap'})

    # Extract data from the table
    if table:
        rows = table.find_all('tr')

        # Initialize data dictionary for DataFrame
        data_dict = {}
        columns = None

        for row in rows:
            cells = row.find_all(['td', 'th'])
            if not columns:
                # Extract column names from the first row (header)
                columns = [cell.text.strip() for cell in cells]
                for col in columns:
                    data_dict[col] = []

            else:
                # Extract data for each column
                for i, cell in enumerate(cells):
                    data_dict[columns[i]].append(cell.text.strip())

        # Create a DataFrame from the data dictionary
        df = pd.DataFrame(data_dict)
        print(df)

    else:
        print("Table not found on the page.")

else:
    print('Failed to fetch the webpage')
```

## Run all tables in loop and save in csv

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = r"https://www.screener.in/company/IRFC/"
response = requests.get(url)

List_df_names = ['Quarterly Results','Profit & Loss','Compounded Sales Growth','Compounded Profit Growth',
                 'Stock Price CAGR','Return on Equity','Balance Sheet','Cash Flows','Ratios','Shareholding Pattern','duplicate_Shareholding Pattern']

if response.status_code == 200:
    html_code = response.text
    soup = BeautifulSoup(html_code, 'html.parser')
    tables = soup.find_all('table')
    dataframes = []

    for idx, table in enumerate(tables):
        df = pd.read_html(str(table))[0]
        dataframes.append(df)

        # Check if the index is within the range of List_df_names
        if idx < len(List_df_names):
            # Assign the name to the DataFrame
            df_name = List_df_names[idx]
            print(f"{df_name} DataFrame:\n{df}\n\n")
            
            # Save each DataFrame to a CSV file with the corresponding name
            df.to_csv(f"{df_name}.csv", index=False)

else:
    print(f"Failed and Status code: {response.status_code}")
```
